```python
import pandas as pd
import seaborn as sns
import sklearn
```

```python
boston = pd.read_csv('https://an-utd-python.s3.us-west-1.amazonaws.com/BostonHousePrices.csv')
boston
```

```python
# 2 attributes -> rooms and distance from downtown
# predict -> value
# tell me the data distribution of Value column
# 2 ways: tabular, graphical way
boston.describe()
```

```python
sns.displot(data = boston, x = "Value")
```

```python
# analyze outliers in the data -> Value column
# use box and whisker plot
sns.boxplot(data = boston, x = "Value")
```

```python
# you may need to consider a separate model for outliers
# pair plot and see if you can uncover any relationships
sns.pairplot(boston)
```

```python
def checkRange(value):
  if ((value > 10) & (value < 30)):
    return "Normal"
  else:
    return "Outlier"
```

```python
checkRange(20)
```

```python
boston["category"] = boston["Value"].apply(lambda x: checkRange(x))
```

```python
boston
```

```python
sns.pairplot(boston, hue = "category")
```

```python
# check if a linear relationship between Value (y-axis) and Rooms (x-axis) makes any sense
# you may want to slice the data based on category is normal or outlier
# check if order = 2 makes better sense
# Repeat this for Value (y-axis) and Distance (x-axis)
plt.figure(figsize = (12, 12))
ax = sns.lmplot(x = "Distance", y = "Value", data = boston[boston["category"] == "Normal"].loc[0:100, :], order = 3)
```

```python
# correlation values
cor = boston.corr()
```

```python
cor # correlation between all pairs 
```

```python
sns.pairplot(boston)
```

```python
# you define the attributes that you are going to use and you define  your predicted variable
# target
boston_normal = boston[boston["category"] == "Normal"]
y = boston_normal["Value"]
# attributes
X = boston_normal[["Rooms", "Distance"]]
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()  # definition of linear model
lm.fit(X, y) # fitting the model on my X and y
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
boston.describe()
```

```python
# before going ahead and creating the model, i should standardize my data (putting them on same scale)
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
X_norm = pd.DataFrame(mm.fit_transform(X), columns = X.columns)
```

```python
X_norm.describe() # the data has been converted to min = 0 and max = 1
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = pd.DataFrame(sc.fit_transform(X), columns = X.columns)
```

```python
X_scaled.describe()
```

```python
# instead of using X (un-normalized data), use X_norm to build the model
lr = linear_model.LinearRegression()
lr.fit(X_norm, y)
```

```python
lr
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y
```

```python
y_pred = lr.predict(X_norm)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y, y_pred)
```

```python
mean_absolute_error(y, y_pred)  # better when i remove outliers
```

```python
r2_score(y, y_pred) # linear model is bad at explaining variance in the data
```
