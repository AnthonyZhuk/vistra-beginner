```python
air_quality
```

```python
# how to handle null values, missing data, inconsistent values
```

```python
df_netflix = pd.read_csv("https://an-utd-python.s3.us-west-1.amazonaws.com/netflix_titles.csv")
df_netflix.head()
```

```python
df_netflix.isna().sum() # how many null values in each column
```

```python
df_netflix.drop('director', axis=1).isna().sum()  # this is throwing away useful information
```

```python
no_director = df_netflix[df_netflix['director'].isnull()].index # get index of rows with missing director
temp = df_netflix.drop(no_director, axis=0)
```

```python
temp.isna().sum()  # throwing away rows where director is null
```

```python
# for text columns, there is no easy way
# for numeric columns, one solution is to replace the nulls with mean
df_netflix[df_netflix['rating'].isnull()]
```

```python
# for text columns, if there are null values, you can replace them with the most common value
df_netflix[df_netflix['rating'].isnull()]
mode = ''.join(df_netflix['rating'].mode())
df_netflix['rating'].fillna(mode, inplace=True)
```

```python
df_netflix.loc[[5989,6827], :]
```

```python
df_netflix['duration'].fillna('0', inplace=True)
```

```python
# visual approaches and tabular approaches to identify outliers in the dataset
df_movie = df_netflix[df_netflix['type']=='Movie']
df_movie['duration']
```

```python
df_movie = df_movie.assign(minute = df_movie['duration'].str.split(expand=True)[0].astype(int))
```

```python
df_movie
```

```python
df_movie['minute'].describe()
```

```python
ax = df_movie['minute'].plot(kind='hist', bins=20)
```

```python
df_movie['minute'].plot(kind='box', vert=False, color='blue',
                        figsize=(10,5))

plt.xlabel('Duration (minutes)')
plt.show()
```

```python
# above is a box plot
# box (25 - 75 percentile and a vertical line at the mean)
# 2 SD above and below the mean
# anything beyond the blue lines can be called outliers
df_movie[(df_movie['minute']<46.5) | (df_movie['minute']>154.5)]
```
